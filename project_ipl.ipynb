## Data Cleaning

```python
import pandas as pd
import numpy as np
```

## Importing raw data

```python
raw_matches = pd.read_csv('matches.csv')
```

```python
raw_deliveries = pd.read_csv('deliveries.csv')
```

## Creating a copy of raw data

```python
matches = raw_matches.copy()
deliveries = raw_deliveries.copy()
```

```python
matches.shape, deliveries.shape
```

## Cleaning Matches

```python
matches.info()
```

## Checking null values and replacing them with appropriate values

```python
matches.isnull().sum()
```

```python
matches['winner'] = matches['winner'].fillna('No Result')
```

```python
matches['city'] = matches['city'].fillna('Unknown')
```

```python
matches['player_of_match'] = matches['player_of_match'].fillna('No Result')
```

```python
matches['umpire1'] = matches['umpire1'].fillna("Unknown")
matches['umpire2'] = matches['umpire2'].fillna("Unknown")
matches['umpire3'] = matches['umpire3'].fillna("Unknown")
```

```python
matches.head()
```

## Giving appropriate datatypes to values of columns

```python
cols = ['city', 'team1', 'team2', 'toss_winner', 'toss_decision', 'result', 'winner', 'player_of_match', 'venue', 'umpire1', 'umpire2', 'umpire3']

matches[cols] = matches[cols].astype(str)
```

```python
matches['date'] = pd.to_datetime(
    matches['date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
matches['date'].isnull().sum()
```

## Standardising Data

```python
team_map = {
    "Mumbai Indians": "MI",
    "Chennai Super Kings": "CSK",
    "Royal Challengers Bangalore": "RCB",
    "Delhi Daredevils": "DC",
    "Delhi Capitals": "DC",
    "Kings XI Punjab": "PBKS",
    "Punjab Kings": "PBKS",
    "Sunrisers Hyderabad": "SRH",
    "Kolkata Knight Riders": "KKR",
    "Rajasthan Royals": "RR",
    "Rising Pune Supergiants": "RPS",
    "Pune Warriors": "RPS",
    "Gujarat Lions": "GL"
}

for col in ['team1', 'team2', 'winner', 'toss_winner']:
    matches[col] = matches[col].replace(team_map)
```

```python
team_map = { 
    "Rising Pune Supergiant": "RPS",
    "Deccan Chargers": "SRH",
    "Kochi Tuskers Kerala": "KTK",
}

for col in ['team1', 'team2', 'winner', 'toss_winner']:
    matches[col] = matches[col].replace(team_map)
```

```python
matches.shape
```

```python
matches = matches[matches['team1'] != matches['team2']]
```

```python
matches['result_type'] = np.where(
    matches['win_by_runs'] > 0, 
    "Won by runs",
    np.where(matches['win_by_wickets'] > 0, "Won by wickets", "No Result")
)
```

## Checking duplicate values

```python
matches.duplicated().sum()
```

## Cleaning deliveries

```python
deliveries.head()
```

```python
deliveries['total_runs'].unique()
```

## Checking null values and replacing them with appropriate values

```python
deliveries.isnull().sum()
```

```python
deliveries.fillna({
    'dismissal_kind': "Not Out",
    'player_dismissed': "None",
    'fielder': "None"
}, inplace=True)
```

## Standardising Data

```python
for col in ['batting_team', 'bowling_team']:
    deliveries[col] = deliveries[col].replace(team_map)
```

```python
deliveries['batsman'] = deliveries['batsman'].str.strip()
deliveries['bowler'] = deliveries['bowler'].str.strip()
```

## Removing duplicate values

```python
deliveries.duplicated().sum()
```

```python
deliveries.drop_duplicates(inplace=True)
```

## Merging Dataframes

```python
ipl = deliveries.merge(
    matches,
    left_on='match_id',
    right_on='id',
    how='left'
)
```

```python
ipl.head()
```

```python
ipl.shape
```

## Post merge data cleaning

```python
ipl.duplicated().sum()
```

```python
ipl.drop(['id'], axis=1, inplace=True)
```

```python
ipl.drop(['team1', 'team2'], axis=1, inplace=True)
```

## Creating advanced features

```python
ipl['is_ball'] = np.where(
    (ipl['wide_runs'] > 0) | (ipl['noball_runs'] > 0),
    0,
    1
)
```

```python
ipl['phase'] = pd.cut(
    ipl['over'],
    bins=[0, 6, 15, 20],
    labels=['Powerplay', 'Middle', 'Death']
)
```

```python
ipl['is_boundary'] = ipl['batsman_runs'].apply(lambda x: 1 if x in [4,6] else 0)
```

```python
ipl.shape, matches.shape, deliveries.shape
```

```python
ipl['strike_rate'] = (ipl['batsman_runs'] / ipl['is_ball']) * 100
```

```python
ipl['strike_rate'] = ipl['strike_rate'].fillna(0)
```

## Cleaned data into csv file

```python
matches.to_csv("matches_cleaned.csv", index=False)
deliveries.to_csv("deliveries_cleaned.csv", index=False)
ipl.to_csv("ipl_cleaned.csv", index=False)
```
